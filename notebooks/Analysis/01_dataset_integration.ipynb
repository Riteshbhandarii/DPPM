```python
# keep only rows scraped during the shared two-week window and with a valid `product_id`.
def clean_window(df):
    df = df.copy()
    df["scrape_date"] = pd.to_datetime(df["scrape_date"], errors="coerce")
    return df.loc[
        df["scrape_date"].between("2026-02-03", "2026-02-18")
        & df["product_id"].notna()
    ]

vw = clean_window(vw)
toyota = clean_window(toyota)
skoda = clean_window(skoda)

df = pd.concat([vw, toyota, skoda], ignore_index=True)
```

## Final cleanup for the merged CSV

This is added after the original code so the merged dataset is easy to use in the next notebook.


```python
# clean text fields so the saved CSV is easy to read later
df_final = df.copy()
text_cols = ['part_name', 'quality_grade', 'oem_number', 'brand', 'model', 'category', 'subcategory']
for col in text_cols:
    df_final[col] = df_final[col].astype('string').str.replace(r'\s+', ' ', regex=True).str.strip()

# remove only exact duplicate rows within the same snapshot
df_final = df_final.drop_duplicates().copy()

print('rows after cleanup:', len(df_final))
print('duplicates removed:', len(df) - len(df_final))
df_final.head()
```
