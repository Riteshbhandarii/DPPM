```python
# keep only rows scraped during the shared two-week window and with a valid `product_id`.
def clean_window(df):
    df = df.copy()
    df["scrape_date"] = pd.to_datetime(df["scrape_date"], errors="coerce")
    return df.loc[
        df["scrape_date"].between("2026-02-03", "2026-02-18")
        & df["product_id"].notna()
    ]

vw = clean_window(vw)
toyota = clean_window(toyota)
skoda = clean_window(skoda)

df = pd.concat([vw, toyota, skoda], ignore_index=True)
```
