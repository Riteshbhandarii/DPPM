# Loading and Merging Snapshot Data from Varaosahaku.fi

## 1) Imports

```python
from pathlib import Path
import pandas as pd
```

## 2) Group CSV Files by Car Brand and Model

```python
root = Path.cwd() if (Path.cwd() / 'README.md').exists() else Path.cwd().parent
all_csv_files = sorted(root.rglob('*.csv'))

golf_files = []
skoda_files = []
toyota_files = []

for csv_file in all_csv_files:
    csv_text = str(csv_file).lower()

    if 'golf' in csv_text or 'e_golf' in csv_text:
        golf_files.append(csv_file)
        continue

    if 'skoda' in csv_text or 'octavia' in csv_text:
        skoda_files.append(csv_file)
        continue

    if 'toyota' in csv_text or 'corolla' in csv_text or 'yaris' in csv_text:
        toyota_files.append(csv_file)
```

## 3) Merge Each Group

```python
golf_frames = []
for csv_file in golf_files:
    df = pd.read_csv(csv_file)
    df['source_file'] = str(csv_file.relative_to(root))
    golf_frames.append(df)
golf_df = pd.concat(golf_frames, ignore_index=True, sort=False)

skoda_frames = []
for csv_file in skoda_files:
    df = pd.read_csv(csv_file)
    df['source_file'] = str(csv_file.relative_to(root))
    skoda_frames.append(df)
skoda_df = pd.concat(skoda_frames, ignore_index=True, sort=False)

toyota_frames = []
for csv_file in toyota_files:
    df = pd.read_csv(csv_file)
    df['source_file'] = str(csv_file.relative_to(root))
    toyota_frames.append(df)
toyota_df = pd.concat(toyota_frames, ignore_index=True, sort=False)
```

## 4) Save the Merged Files

```python
output_dir = root / 'datasets' / 'merged'
output_dir.mkdir(parents=True, exist_ok=True)

golf_df.to_csv(output_dir / 'dppm_vw_golf_e_golf.csv', index=False)
skoda_df.to_csv(output_dir / 'dppm_skoda_octavia.csv', index=False)
toyota_df.to_csv(output_dir / 'dppm_toyota_corolla_yaris.csv', index=False)
```
